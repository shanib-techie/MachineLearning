```python
import pandas as pd
import numpy as np
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_absolute_error,mean_squared_error,root_mean_squared_error

df = pd.read_csv("C:\\Users\\Admin\\OneDrive\\Desktop\\MachineLearning\\ML_practise\\student_score.csv")

X = df[["Hours"]]
y = df["Score"]

print(df.describe())
model = LinearRegression()
model.fit(X,y)
num_of_diff_hours = int(input("ENTER HOW MUCH DIFFERENT HOURS YOU WANT TO WIRTE :"))
list = []
for i in range(1,num_of_diff_hours+1):
    hours = int(input(f"ENTER THE VALUES OF{i} EXAMPLE HOURS : "))
    list.append(hours)
# # predict_score = model.predict(list)  ##yeh nhi calega  kyuki hum ek baar m multiple value dekh rhe h
list_2d = [[i] for i in list]
predict_score = model.predict(list_2d)
for i, pred in zip(list, predict_score):
    print(f"If you study {i} hours score is {pred}")
```
