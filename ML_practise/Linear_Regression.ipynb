```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error

# -----------------------------
# STEP 1: Data load karo
# -----------------------------
df = pd.read_csv("student_score.csv")

# X = input (hours), y = output (score)
X = df[["Hours"]]
y = df["Score"]

# -----------------------------
# STEP 2: Train-Test Split
# -----------------------------
# 80% training, 20% testing
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)

# -----------------------------
# STEP 3: Model train karo
# -----------------------------
model = LinearRegression()
model.fit(X_train, y_train)

# -----------------------------
# STEP 4: Prediction karo (test data pe)
# -----------------------------
y_pred = model.predict(X_test)

# -----------------------------
# STEP 5: Error calculate karo (MSE, RMSE)
# -----------------------------
mse = mean_squared_error(y_test, y_pred)
rmse = np.sqrt(mse)

print("MSE:", mse)
print("RMSE:", rmse)

# -----------------------------
# STEP 6: Graph plot karo
# -----------------------------

# 1. Original data (scatter)
plt.scatter(X, y)

# 2. Best fit line (model ne jo seekha)
X_sorted = np.sort(X, axis=0)
y_line = model.predict(X_sorted)
plt.plot(X_sorted, y_line)

# 3. Test predictions (highlight)
plt.scatter(X_test, y_pred)

plt.xlabel("Study Hours")
plt.ylabel("Score")
plt.title("Student Score Prediction")

plt.show()

# -----------------------------
# STEP 7: New prediction (user input)
# -----------------------------
hours = float(input("Enter study hours: "))
pred = model.predict([[hours]])

print(f"If you study {hours} hours, expected score ≈ {pred[0]}")
```

```python
import pandas as pd
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error

# Step 1: Data
df = pd.DataFrame({
    "experience": [1, 2, 3, 4, 5],
    "level": ["low", "medium", "medium", "high", "high"],
    "salary": [20000, 30000, 35000, 50000, 55000]
})

# Step 2: String → Number (One-Hot Encoding)
df_encoded = pd.get_dummies(df, columns=["level"])

# Step 3: Features (X) and Target (y)
X = df_encoded.drop("salary", axis=1)
y = df_encoded["salary"]

# Step 4: Model
model = LinearRegression()
model.fit(X, y)

# Step 5: Prediction
y_pred = model.predict(X)

# Step 6: Error
mse = mean_squared_error(y, y_pred)

print("Encoded Data:\n", df_encoded)
print("\nPredictions:", y_pred)
print("\nMSE:", mse)
```
