```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error

# -----------------------------
# STEP 1: Data load karo
# -----------------------------
df = pd.read_csv("student_score.csv")

# X = input (hours), y = output (score)
X = df[["Hours"]]
y = df["Score"]

# -----------------------------
# STEP 2: Train-Test Split
# -----------------------------
# 80% training, 20% testing
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)

# -----------------------------
# STEP 3: Model train karo
# -----------------------------
model = LinearRegression()
model.fit(X_train, y_train)

# -----------------------------
# STEP 4: Prediction karo (test data pe)
# -----------------------------
y_pred = model.predict(X_test)

# -----------------------------
# STEP 5: Error calculate karo (MSE, RMSE)
# -----------------------------
mse = mean_squared_error(y_test, y_pred)
rmse = np.sqrt(mse)

print("MSE:", mse)
print("RMSE:", rmse)

# -----------------------------
# STEP 6: Graph plot karo
# -----------------------------

# 1. Original data (scatter)
plt.scatter(X, y)

# 2. Best fit line (model ne jo seekha)
X_sorted = np.sort(X, axis=0)
y_line = model.predict(X_sorted)
plt.plot(X_sorted, y_line)

# 3. Test predictions (highlight)
plt.scatter(X_test, y_pred)

plt.xlabel("Study Hours")
plt.ylabel("Score")
plt.title("Student Score Prediction")

plt.show()

# -----------------------------
# STEP 7: New prediction (user input)
# -----------------------------
hours = float(input("Enter study hours: "))
pred = model.predict([[hours]])

print(f"If you study {hours} hours, expected score ≈ {pred[0]}")
```

```python
import pandas as pd
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error

# Step 1: Data
df = pd.DataFrame({
    "experience": [1, 2, 3, 4, 5],
    "level": ["low", "medium", "medium", "high", "high"],
    "salary": [20000, 30000, 35000, 50000, 55000]
})

# Step 2: String → Number (One-Hot Encoding)
df_encoded = pd.get_dummies(df, columns=["level"])

# Step 3: Features (X) and Target (y)
X = df_encoded.drop("salary", axis=1)
y = df_encoded["salary"]

# Step 4: Model
model = LinearRegression()
model.fit(X, y)

# Step 5: Prediction
y_pred = model.predict(X)

# Step 6: Error
mse = mean_squared_error(y, y_pred)

print("Encoded Data:\n", df_encoded)
print("\nPredictions:", y_pred)
print("\nMSE:", mse)
```

```python
import pandas as pd
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error

# -------------------------------
# STEP 1: Simple data bana rahe hain
# hours = kitna padha
# location = kaha padha (string)
# marks = result (target)
# -------------------------------

df = pd.DataFrame({
    "hours": [1,2,3,4,5,6,2,3,4,5,6,7],
    "location": [
        "home","library","school","tutor",
        "online","home","library","school",
        "tutor","online","home","library"
    ],
    "marks": [30,45,50,55,60,70,40,52,58,65,75,80]
})

# -------------------------------
# STEP 2: String → Number (One-Hot Encoding)
# location ko machine ke liye numbers me convert kar rahe hain
# -------------------------------

df_encoded = pd.get_dummies(df, columns=["location"])

# -------------------------------
# STEP 3: X (input) aur y (output)
# X = features (hours + location columns)
# y = marks
# -------------------------------

X = df_encoded.drop("marks", axis=1)
y = df_encoded["marks"]


# -------------------------------
# STEP 5: Train-Test Split
# data ka kuch part training ke liye, kuch testing ke liye
# random_state fix rakha taaki result same aaye
# -------------------------------

X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.3, random_state=42
)

# -------------------------------
# STEP 6: Model train (Linear Regression)
# model pattern seekhega
# -------------------------------

model = LinearRegression()
model.fit(X_train, y_train)

# -------------------------------
# STEP 7: Prediction
# test data pe marks predict karega
# -------------------------------

y_pred = model.predict(X_test)

# -------------------------------
# STEP 8: Error (MSE)
# actual vs predicted ka difference
# -------------------------------

mse = mean_squared_error(y_test, y_pred)
print("Test Index:", y_test.index.tolist())

print("=== Actual Marks ===")
print(list(y_test))

print("\n=== Predicted Marks ===")
print(list(y_pred))

print("\nMSE:", mse)
print("\n--- Prediction Mode ---")

while True:
    hours = int(input("Enter hours: "))
    location = input("Enter location (home/library/school/tutor/online): ")

    # simple input ko same format me convert kar rahe hain
    new_data = pd.DataFrame({
        "hours": [hours],
        "location_home": [1 if location == "home" else 0],
        "location_library": [1 if location == "library" else 0],
        "location_online": [1 if location == "online" else 0],
        "location_school": [1 if location == "school" else 0],
        "location_tutor": [1 if location == "tutor" else 0]
    })

    # prediction
    pred = model.predict(new_data)

    print(f"Predicted Marks: if someone study : {hours} and location is : {location}", pred[0])

    # loop band karne ke liye
    ch = input("Continue? (y/n): ")
    if ch != 'y':
        break
```
