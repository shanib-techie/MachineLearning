```python
# Linear Regression  => work with numbers give the answer of next pridiction
from sklearn.linear_model import LinearRegression
X = [[1],[2],[3],[4],[5],[6]]
y = [20,45,60,70,100,140]
model = LinearRegression()
model.fit(X,y)
hour = int(input("ENTER HOW MUCH HOURS YOU STUDIES"))
pridict_marks = model.predict([[hour]])
print(f"based on the input {hour} you may be score around {pridict_marks}")
```

```python
##classification [LogisticRegression] give the answer in yes no 
from sklearn.linear_model import LogisticRegression
X = [[1],[2],[3],[4],[5]]
y = [0,0,0,1,1,] ##result 0 = f 1 = p
model = LogisticRegression()
model.fit(X,y)

hour = float(input("ENTER HOW MUCH HOURS YOU STUDIES"))

result = model.predict([[hour]])[0]

if result == 1:
    print(f"based on the input {hour} you may be pass")
else:
    print(f"based on the input {hour} you may be fail")



```

```python
from sklearn.neighbors import KNeighborsClassifier
X = [
    [180,7],
    [200,7.5],
    [250,8],
    [300,8.5],
    [330,9],
    [360,9.5]
]
y = [0,0,0,1,1,1]##0 =apple 1 = orange
model = KNeighborsClassifier(n_neighbors=3)
model.fit(X,y)
weight = float(input("ENTER THE WEIGHT OF FRUIT :"))
size = float(input("ENTER THE SIZE OF FRUIT : "))
prediction = model.predict([[weight,size]])[0]
if prediction == 0:
    print("this is likely apple")
else:
    print("this is likey orange")    
```

```python
from sklearn.tree import DecisionTreeClassifier
X = [
    [7,2],##size  colorshade
    [8,3],
    [9,8],
    [10,9]
]
y = [0,0,1,1]  #0 apple 1 oranger
model = DecisionTreeClassifier()
model.fit(X,y)

size = float(input("ENTER THE SIZE OF FRUIT : "))
shade = int(input("ENTER THE COLOR SHADE (1-10) : "))

result=model.predict([[size,shade]])[0]

if result == 0:
    print("THIS IS LIKELY TO APPLE ")
else:
    print("THIS IS LIKELY TO ORANGE")
```
