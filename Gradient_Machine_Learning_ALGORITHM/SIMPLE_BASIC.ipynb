```python
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import GradientBoostingRegressor
import pandas as pd

# ----------------------------
# Dataset
# ----------------------------

data = {
    "Age": [20, 25, 30, 35, 40, 45, 50],
    "BMI": [22, 24, 26, 28, 30, 32, 34],
    "Charges": [2000, 3000, 5000, 7000, 10000, 14000, 18000]
}

df = pd.DataFrame(data)

X = df[["Age", "BMI"]]
y = df["Charges"]

# ----------------------------
# Gradient Boosting Model
# ----------------------------

model = GradientBoostingRegressor(
    n_estimators=3,      # Sirf 3 trees
    learning_rate=1,     # Full correction
    max_depth=2,
    random_state=42
)

model.fit(X, y)

# ----------------------------
# Prediction
# ----------------------------

prediction = model.predict([[38, 29]])

print("Predicted Charges :", prediction[0])
```
