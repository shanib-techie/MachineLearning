```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.metrics import root_mean_squared_error , r2_score
```

```python
df = pd.read_csv("C:\\Users\\Admin\\OneDrive\\Desktop\\MachineLearning\\ML_practise\\fear_greed_index.csv")
print(df.head(10))
print(df["classification"].unique())
```

```python
df_encoded = pd.get_dummies(df,columns=["classification"])
print(df_encoded)
```

```python
X = df_encoded.drop(columns=["timestamp","date","value"])
y = df_encoded["value"]

X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.1)

model = LinearRegression()
model.fit(X_train,y_train)
y_pridict = model.predict(X_test)

print("PRIDICT VALUE OF :",y_pridict)
print("\n ACTUAL VALUE : ",y_test)
print("R2 SCORE : ",r2_score(y_test,y_pridict))
print("\nroot_mean_squared_error : ",root_mean_squared_error(y_pridict,y_test))
```
